```python
import requests
import json
import pandas as pd 
import time
from datetime import timedelta, datetime
import sys
```

### Definición del token y la cabecera

```python
token = "" # Rellenar con token generado en GitHub
header = {'Authorization': 'token %s' % token}
```

### Función de generación de histórico de repositorios

```python
cols = ["num_repos", "date"]
```

```python
res_df = pd.DataFrame(columns=cols)

date = datetime(2016, 1, 1)
limit_date = datetime(2018, 12, 31)

while(date <= limit_date):
    print("\n\n -- PROCESANDO FECHA {} --".format(date.strftime("%Y-%m-%d")))
    
    query = "?q=created:" + date.strftime("%Y-%m-%d")

    response = requests.get('https://api.github.com/search/repositories' + query, headers=header)
    response_json = response.json()
    print(response.url)
    total_count = response_json["total_count"]
    
    temp_df = pd.DataFrame(columns=cols)
    temp_df.loc[0] = [total_count, date.strftime("%Y-%m-%d")]
    
    res_df = res_df.append(temp_df)
    print("Tamaño actual dataframe: {}".format(res_df.shape[0]))
        
    print("Esperando 2 segundos...")
    time.sleep(2) #Espera de 2 segundos
        
    date += timedelta(days = 1)
    
res_df.to_csv(r'historicos/num_repos_'+ date.strftime("%Y") + '-' + limit_date.strftime("%Y") + '.csv', index = False, encoding="utf-8-sig")
```

### Unión de los dos dataframes generados

```python
repos_2016_2018 = pd.read_csv('historicos/num_repos_2016-2018.csv')
repos_2019_2021 = pd.read_csv('historicos/num_repos_2019-2021.csv')

repos_2016_2021 = repos_2016_2018.append(repos_2019_2021)

repos_2016_2021.to_csv(r'historicos/historico_repositorios_2016_2021.csv', index = False, encoding="utf-8-sig")
```
